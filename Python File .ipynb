```python
# (2.a) Load CSV in R by skipping second row.

import pandas as pd

df= pd.read_csv("IMDB_data.csv", sep = ',', encoding = 'Latin-1')

df=df.drop(df.index[1])
```

```python
df
```

```python
# (2.b) Extract the unique genres and its count and store in data frame with index key. 

unique_genre = df['Genre'].value_counts()
unique_genre_df = pd.DataFrame(unique_genre.to_frame().reset_index())
unique_genre_df
```

```python
# (2.c) Convert the required data types.

df['imdbVotes'] = pd.to_numeric(df['imdbVotes'],errors='coerce')
df['imdbRating'] = pd.to_numeric(df['imdbRating'],errors='coerce')
df
```

```python
# (2.d) Sort the genre by its name.

df1 = df.sort_values(by=['Genre'])
df2 = pd.DataFrame(df1.Genre)
```
